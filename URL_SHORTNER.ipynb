```python
import pyshorteners

def urlshortner(url):
  shortner=pyshorteners.Shortener()
  s=shortner.tinyurl.short(url)
  return s

url="https://www.toppr.com/guides/computer-science/introduction-to-python/getting-started-with-python/python/"
print(urlshortner(url))
```
